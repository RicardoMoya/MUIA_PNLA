```python
import spacy

# Cargamos el modelo de español
nlp = spacy.load("es_core_news_sm")

# Texto que queremos analizar 
texto = "El Real Madrid jugará la final en Londres."

# Procesamos el texto con spaCy
doc = nlp(texto)

# Obtenemos los tokens
tokens = [token.text for token in doc]
print(f"Texto tokenizado: {tokens}")

# Mostramos información de cada token
print(f"\n{'Texto':<15} {'Lema':<15} {'Categoría':<12} {'Stopword'}")
print("-" * 60)
for token in doc:
    print(f"{token.text:<15} {token.lemma_:<15} {token.pos_:<12} {token.is_stop}")

# Mostramos las entidades detectadas 
print("\nENTIDADES")
print("-" * 20)

for entidad in doc.ents:
    print(f"{entidad.text} → {entidad.label_}")
```
