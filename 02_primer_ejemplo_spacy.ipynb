# 02. Primer ejemplo práctico con spaCy

En este notebook vamos a ver un primer ejemplo práctico con **spaCy**, la librería de Python utilizada para procesar textos en lenguaje natural.

El objetivo es entender qué información podemos obtener a partir de una frase sencilla.

A lo largo del notebook veremos cómo:

* Cargar un modelo de español de spaCy.
* Procesar un texto.
* Obtener sus tokens.
* Consultar el lema de cada palabra.
* Identificar la categoría gramatical de cada token.
* Detectar stopwords.
* Extraer entidades nombradas.

spaCy organiza el procesamiento del texto mediante un objeto llamado `nlp`. Este objeto contiene el pipeline lingüístico, es decir, la secuencia de pasos que se aplican al texto. Cuando pasamos una frase a `nlp`, spaCy devuelve un objeto de tipo `Doc`.

El objeto `Doc` representa el texto procesado. Dentro de él se almacenan los tokens y la información lingüística calculada por el modelo, como lemas, categorías gramaticales o entidades nombradas.


````python
import spacy

nlp = spacy.load("es_core_news_sm")
````

## 1. Texto de ejemplo

Vamos a trabajar con una frase sencilla relacionada con el fútbol:

```text
El Real Madrid jugará la final en Londres.
```

Esta frase nos permite ver varios elementos interesantes. Por ejemplo, contiene una entidad formada por varias palabras (`Real Madrid`), una forma verbal conjugada (`jugará`) y un lugar (`Londres`).


````python
texto = "El Real Madrid jugará la final en Londres."

doc = nlp(texto)
````

## 2. Tokenización

El primer resultado que podemos consultar es la **tokenización**. Tokenizar consiste en dividir el texto en unidades más pequeñas llamadas **tokens**.

Un token puede ser una palabra, un signo de puntuación, un número o una unidad lingüística detectada por el modelo.


````python
tokens = [token.text for token in doc]

print(f"Texto tokenizado: {tokens}")
````

## 3. Información lingüística de cada token

Cada token del objeto `Doc` contiene diferentes atributos. En este ejemplo vamos a consultar cuatro:

* `token.text`: texto original del token.
* `token.lemma_`: lema o forma base de la palabra.
* `token.pos_`: categoría gramatical del token.
* `token.is_stop`: indica si el token es una stopword.

Por ejemplo, la forma verbal `jugará` debería tener como lema `jugar`.


````python
print(f"{'Texto':<15} {'Lema':<15} {'Categoría':<12} {'Stopword'}")
print("-" * 60)

for token in doc:
    print(f"{token.text:<15} {token.lemma_:<15} {token.pos_:<12} {token.is_stop}")
    
````

## 4. Entidades nombradas

Las entidades detectadas por spaCy se almacenan en el atributo `doc.ents` del objeto `Doc`.

Cada elemento de `doc.ents` es un objeto de tipo `Span` que representa una secuencia de uno o varios tokens identificados por el modelo.

En este ejemplo vamos a consultar dos atributos:

* `entidad.text`: devuelve el texto original de la entidad detectada.
* `entidad.label_`: devuelve la etiqueta asignada por el modelo.

Por ejemplo, si spaCy detecta la entidad `Real Madrid`, el atributo `text` devolverá el texto `"Real Madrid"` y el atributo `label_` devolverá la categoría asignada por el modelo.

A continuación recorreremos el atributo `doc.ents` para mostrar las entidades detectadas junto con su etiqueta correspondiente.


````python
print("ENTIDADES")
print("-" * 30)

for entidad in doc.ents:
    print(f"{entidad.text} → {entidad.label_}")
````

## 5. Conclusión

En este primer ejemplo hemos comprobado cómo spaCy transforma un texto normal en un objeto `Doc` enriquecido con información lingüística.

A partir de una sola frase hemos podido obtener sus tokens, consultar el lema de cada palabra, identificar categorías gramaticales, detectar stopwords y extraer entidades nombradas.

Si todas las celdas se ejecutan correctamente, significa que spaCy está instalado, que el modelo `es_core_news_sm` se ha cargado sin problemas y que el notebook está utilizando el entorno de Python adecuado.
